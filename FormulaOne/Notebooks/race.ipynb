```python
import numpy as np
import pandas as pd
```

```python
race=pd.read_csv('F1_Race_Results.csv',sep=',')
race
```

```python
race['Date']=pd.to_datetime(race['Date'],format='mixed')
race['Year']=race['Date'].dt.year
race.Year
```

```python
race.columns
```

### 1.1 Most Successful Drivers by wins

```python
race['Winner'].value_counts()
```

### 1.2 Finding most successful car with most time winner(in this case Lewis Hamilton)

```python
most_driver_wins=race[race['Winner'] == 'Lewis Hamilton HAM']
most_driver_wins
```

```python
car_wins=most_driver_wins['Car'].value_counts()
print(car_wins)
```

### 1.2 Best Car(By Wins)

```python
race['Car'].value_counts()
```

### 1.3 Most Dominant periods

```python
dominance=race.groupby('Year')['Winner'].value_counts().sort_values(ascending=False)
dominance
```

```python
# Conversion to DataFrame
domination = dominance.reset_index()
domination.columns = ['Year', 'Winner', 'Times']
domination
```

```python
domination[domination['Times']>10]
```

### 1.4. Identify eras: Ferrari era, Red Bull era, Mercedes era

```python
dominance_car=race.groupby(['Year','Car'])['Winner'].value_counts().sort_values(ascending=False)
dominance_car
```

```python
# Conversion to DataFrame
domination_car = dominance_car.reset_index()
domination_car.columns = ['Year', 'Car', 'Winner', 'Times']
domination_car
```

### 2.1 Grand Prix where a single constructor wins repeatedly

```python
race['Grand Prix'].value_counts()
```

```python
dominance_car_gp=race.groupby(['Car','Grand Prix'])['Winner'].value_counts().sort_values(ascending=False)
dominance_car_gp
```

```python
# Conversion to DataFrame
domination_car_gp = dominance_car_gp.reset_index()
domination_car_gp.columns = ['Car', 'Grand Prix', 'Winner', 'Times']
domination_car_gp[domination_car_gp['Times']>5]
```

```python
race['Time'] = race['Time'].apply(
    lambda x: 0 if pd.isna(x) else
    sum(int(v) * m for v, m in zip(str(x).replace('.', ':').split(':'), [3600000, 60000, 1000, 1]))
)
```

```python
race
```

```python
race['SR']=(race['Time']/race['Laps'])
race['SR']
```

```python
high_speed=race[race['Laps']>50].sort_values(by='SR',ascending=False).head(10)
high_speed
```

```python
speedy_laps=race.groupby(['Winner','Laps','Time','Grand Prix'])['SR'].value_counts().sort_values(ascending=False)
speedy_laps
```

```python
# Conversion to DataFrame
speedy_laps = speedy_laps.reset_index()
speedy_laps.columns = ['Winner', 'Laps', 'Time_Total', 'Grand Prix', 'SR', 'Times']
speedy_laps[speedy_laps['Laps']>60]
```
